 # What is the best way to handle these missing values and why ?
 
 ##  Best way to handle missing value
- Employee_Name : Drop because its Name 
- Salary : mean Best Way to take the avg for to support the Distribution
- DOB  : MOD  Because its "Date" Type
- Sex : MOD  Because its Categorical with specfic categroical
- DateofTermination : Will replace null values with "Still_Working"
- ManagerID : Drop  Same as Employee name we cant replace it so Will use drop
- Absences : MEDIAN  to avoid outlier with average 

# Apply method to the column

```python
data.dropna(axis=0, subset=['Employee_Name'],inplace=True)
data.dropna(axis=0, subset=['ManagerID'],inplace=True)
```

```python
data.fillna({"Salary": data.Salary.mean(),
            "DOB": data.DOB.mod,
            "Sex":data.Sex.mod,
            "Absences":data.Absences.median()},
           inplace=True)
data.DateofTermination.replace(np.nan,"Still_Working",inplace=True)
```
